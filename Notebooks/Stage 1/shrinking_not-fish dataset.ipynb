# This code was typed to shrink the dataset, by picking 13k pictures

```python
import os
import random
import shutil
```

```python
source_dir = "C:\\Users\\Yazeed\\Desktop\\W\\archive (2)\\Fish-NotFish_Species_Dataset\\Training_Set_Species_Dataset\\Not_Fish"
target_dir = "C:\\Users\\Yazeed\\Desktop\\W\\archive (2)\\Fish-NotFish_Species_Dataset\\Training_Set_Species_Dataset\\Not_Fish_Shrunk"
```

```python
Images = os.listdir(source_dir)
random.shuffle(Images)
```

```python
selected_images = Images[:13000]                #select only 13000 photo (not-fish)
for img in selected_images:                     
    src_path = os.path.join(source_dir, img)
    dst_path = os.path.join(target_dir, img)
    shutil.copyfile(src_path, dst_path)         # copy the photo from the source to the destination (target directory)
```
