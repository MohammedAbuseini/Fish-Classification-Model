# This code was typed to split the 13k picked pictures to (train, test, val)

```python
import os
import random
import shutil
```

```python
SOURCE_DIR = "C:\\Users\\Yazeed\\Desktop\\W\\Not_Fish_Shrunk"  
TARGET_DIR = "C:\\Users\\Yazeed\\Desktop\\W"


#Manualy calculated the data distrubution, and it was as follows

splits = {
    "train": 0.65,
    "val": 0.20,
    "test": 0.15
}

# Create folders
for split in splits:
    os.makedirs(os.path.join(TARGET_DIR, split, "not_fish"), exist_ok=True)

images = os.listdir(SOURCE_DIR)
random.shuffle(images)          #To reasure the model won't be biased

#Calculating the number for each split

total = len(images)
train_end = int(splits["train"] * total)
val_end = train_end + int(splits["val"] * total)

split_map = {
    "train": images[:train_end],        # from 0 to 65%             from all the dataset
    "val": images[train_end:val_end],   # from 65% to (65+20)%      from all the dataset which equals to 20%
    "test": images[val_end:]            # the rest goes to test,    which is 15%
}

# Copy images
for split, imgs in split_map.items():
    for img in imgs:
        shutil.copy(
            os.path.join(SOURCE_DIR, img),
            os.path.join(TARGET_DIR, split, "not_fish", img)
        )
```
